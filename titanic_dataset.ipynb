```python
guess_ages = np.zeros((2,3))
guess_ages
```

```python
for dataset in combine:
    for i in range(0, 2):
        for j in range(0, 3):
            guess_df = dataset[(dataset['Sex'] == i) & \
                                  (dataset['Pclass'] == j+1)]['Age'].dropna()

            # age_mean = guess_df.mean()
            # age_std = guess_df.std()
            # age_guess = rnd.uniform(age_mean - age_std, age_mean + age_std)

            age_guess = guess_df.median()

            # Convert random age float to nearest .5 age
            guess_ages[i,j] = int( age_guess/0.5 + 0.5 ) * 0.5
            
    for i in range(0, 2):
        for j in range(0, 3):
            dataset.loc[ (dataset.Age.isnull()) & (dataset.Sex == i) & (dataset.Pclass == j+1),\
                    'Age'] = guess_ages[i,j]

    dataset['Age'] = dataset['Age'].astype(int)

train_df.head()
```

```python
#Random Forest 

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
acc_random_forest
```

```python
#Naive Bayes 
gaussian = GaussianNB()
gaussian.fit(X_train, Y_train)
Y_pred = gaussian.predict(X_test)
acc_gaussian = round(gaussian.score(X_train, Y_train) * 100, 2)
acc_gaussian
```

```python
#Perceptron 

perceptron = Perceptron()
perceptron.fit(X_train, Y_train)
Y_pred = perceptron.predict(X_test)
acc_perceptron = round(perceptron.score(X_train, Y_train) * 100, 2)
acc_perceptron
```

```python
# Stochastic Gradient Descent

sgd = SGDClassifier()
sgd.fit(X_train, Y_train)
Y_pred = sgd.predict(X_test)
acc_sgd = round(sgd.score(X_train, Y_train) * 100, 2)
acc_sgd
```

```python
#Linear SVC 

linear_svc = LinearSVC()
linear_svc.fit(X_train, Y_train)
Y_pred = linear_svc.predict(X_test)
acc_linear_svc = round(linear_svc.score(X_train, Y_train) * 100, 2)
acc_linear_svc
```

```python
#Decision Tree

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = round(decision_tree.score(X_train, Y_train) * 100, 2)
acc_decision_tree
```

Using Over - Sampling technique (SMOTE) to increase the accuracy of the models. 

- Checking for class imbalance
- Synthetic oversampling of the minority class
- Re-running the models to check for accuracies 

The model accuracies decreased after applying the SMOTE technique. 

```python
# Checking for class imbalance
train_df['Survived'].value_counts(normalize=True)
```

```python
# Importing SMOTE
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Splitting the data into features and target
X = train_df.drop("Survived", axis=1)
y = train_df["Survived"]

# Apply SMOTE (synthetic oversampling of minority class)
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)

# Check the distribution of classes after resampling
print("Before SMOTE:", y.value_counts(normalize=True))
print("After SMOTE:", pd.Series(y_res).value_counts(normalize=True))
```

```python
#Support Vector Machine 
svc = SVC()
svc.fit(X_res, y_res)
Y_pred = svc.predict(X_test)
acc_svc = round(svc.score(X_res, y_res) * 100, 2)
acc_svc


#KNN - K-Nearest Neighbour algorithm 
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_res, y_res)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_res, y_res) * 100, 2)
acc_knn


#Logistic Regression 
logreg = LogisticRegression()
logreg.fit(X_res, y_res)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_res, y_res) * 100, 2)
acc_log


#Random Forest 
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_res, y_res)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_res, y_res)
acc_random_forest = round(random_forest.score(X_res, y_res) * 100, 2)
acc_random_forest


#Naive Bayes 
gaussian = GaussianNB()
gaussian.fit(X_res, y_res)
Y_pred = gaussian.predict(X_test)
acc_gaussian = round(gaussian.score(X_res, y_res) * 100, 2)
acc_gaussian


#Perceptron 
perceptron = Perceptron()
perceptron.fit(X_res, y_res)
Y_pred = perceptron.predict(X_test)
acc_perceptron = round(perceptron.score(X_res, y_res) * 100, 2)
acc_perceptron


# Stochastic Gradient Descent
sgd = SGDClassifier()
sgd.fit(X_res, y_res)
Y_pred = sgd.predict(X_test)
acc_sgd = round(sgd.score(X_res, y_res) * 100, 2)
acc_sgd


#Linear SVC 
linear_svc = LinearSVC()
linear_svc.fit(X_res, y_res)
Y_pred = linear_svc.predict(X_test)
acc_linear_svc = round(linear_svc.score(X_res, y_res) * 100, 2)
acc_linear_svc

#Decision Tree
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_res, y_res)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = round(decision_tree.score(X_res, y_res) * 100, 2)
acc_decision_tree
```

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 
              'Random Forest', 'Naive Bayes', 'Perceptron', 
              'Stochastic Gradient Decent', 'Linear SVC', 
              'Decision Tree'],
    'Score': [acc_svc, acc_knn, acc_log, 
              acc_random_forest, acc_gaussian, acc_perceptron, 
              acc_sgd, acc_linear_svc, acc_decision_tree]})
models.sort_values(by='Score', ascending=False)
```
